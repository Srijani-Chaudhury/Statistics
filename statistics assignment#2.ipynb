```python
def batsmanruns(batsman,matches):
    ipl_data=pd.read_csv('deliveries.csv')
    new_data=ipl_data.groupby(['match_id','batsman'])['batsman_runs'].sum().reset_index()
    d=new_data[new_data['batsman']==batsman].sort_values('match_id').iloc[:matches]['batsman_runs'].sum()
    print("{}(Runs scored by {} after {} matches)".format(d,batsman,matches))
batsmanruns('V Kohli',18)  
```

```python
def ipldata(batsman1,batsman2):    
    ipl_data.head()
    ipl1=ipl_data['batsman']==batsman1
    ipl2=ipl_data['batsman']==batsman2
    ipl=ipl_data[ipl1 | ipl2]
    #sns.FacetGrid(ipl,hue='batsman',height=6).map(sns.distplot,'batsman_runs').add_legend()
    iplnew=ipl.groupby(['match_id','batsman'])['batsman_runs'].sum().reset_index()
    sns.FacetGrid(iplnew,hue='batsman',height=6).map(sns.distplot,'batsman_runs').add_legend()
ipldata('DA Warner','V Kohli')    
```

```python
def CountFrequency(my_list): 
      
    # Creating an empty dictionary  
    l=[]
    freq = {} 
    for items in my_list: 
        freq[items] = my_list.count(items) 
      
    for key, value in freq.items(): 
        l.append(value)
    print(l)    
my_list =[1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2] 
CountFrequency(my_list)         
        
```

```python

def plotpdfcdf(my_list):
    l=[]
    s=0
    counts,data=np.histogram(my_list,bins=5)
    pdf=counts/sum(counts)
    for i in counts:
        s=s+i
        l.append(s)
    print(l)
    cdf=l/sum(counts)
    plt.plot(data[1:],pdf)
    plt.plot(data[1:],cdf)
```
